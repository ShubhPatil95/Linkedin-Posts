# Logging In Python

### Lets do some experiment to understand need of logging library. We have a below function to check if string is valid or not. Now we want to know that due to which condition(condition 1 or 2 or 3) string is invalid. Then what could be your approach?  

```python
#String is invalid for below 3 conditions and otherswise it is valid
# condition 1 : if space is there in string
# condition 2 : if @ symbol in string
# condition 3 : if comma in string

def string_valid(text):
    if " " in text:
        return "Invalid"
    elif "@" in text:
        return "Invalid"
    elif "," in text:
        return "Invalid"
    else:
        return "Valid"

text="Tesing String"

string_valid(text)
```

### Lets try to replace print statements with logger and check if detail is recorded into log file

```python
#importing module
import logging

#Create and configure logger(we can customize the format as per our need, filemode w=write and default is a=append)
logging.basicConfig(filename="LogFile.log",format='%(asctime)s %(levelname)s %(message)s',filemode='a')

#Creating an object
logger=logging.getLogger()

#Setting the threshold of logger to DEBUG (above DEBUGG will be logged only)
logger.setLevel(logging.DEBUG)

def string_valid(text):
    if " " in text:
        logger.debug("space found in text")
#         print("space found in text")
        return "Invalid"
    elif "@" in text:
        logger.debug("@ found in text")
#         print("@ found in text")
        return "Invalid"
    elif "," in text:
        logger.debug("comma found in text")
#         print("comma found in text")
        return "Invalid"
    else:
        logger.info("all conditions are passed")
        return "Valid"

text="Tesing@tring"
string_valid(text)
```

### This is how we can create the log file and every time you will run a function with different string inputs and log will be added accordingly

# The above all can be written as below by creating loger

```python
import logging

# Gets or creates a logger
logger = logging.getLogger()  

# set log level
logger.setLevel(logging.DEBUG)

# define file handler and set formatter
file_handler = logging.FileHandler('logfile.log')
formatter    = logging.Formatter('%(asctime)s : %(levelname)s : %(message)s')
file_handler.setFormatter(formatter)

# add file handler to logger
logger.addHandler(file_handler)

def string_valid(text):
    if " " in text:
        logger.debug("space found in text")
#         print("space found in text")
        return "Invalid"
    elif "@" in text:
        logger.debug("@ found in text")
#         print("@ found in text")
        return "Invalid"
    elif "," in text:
        logger.debug("comma found in text")
#         print("comma found in text")
        return "Invalid"
    else:
        logger.info("all conditions are passed")
        return "Valid"

text="Tesing@tring"
string_valid(text)
```

```python
## Below code will generate new file of last day with date at suffix 

import logging
import logging.handlers
print("---------- Creating Logger ----------")
log_file_name = 'parsing.log'
logging_level = 'DEBUG'
logger=logging.getLogger()
formatter = logging.Formatter('%(asctime)s %(funcName)s %(levelname)s %(message)s')
handler = logging.handlers.TimedRotatingFileHandler(log_file_name, when="midnight", interval=1, backupCount=0)
handler.setFormatter(formatter)
logger = logging.getLogger("Test.log") # or pass string to give it a name
logger.addHandler(handler)
logger.setLevel(logging_level)
handler.namer = lambda name: name.replace(".log", "") + ".log"
print("---------- Logger Created ----------")
```
